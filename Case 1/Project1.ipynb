## Description

1. ***Invoice:*** Номер счета. 6-значный целочисленный номер, уникально присваиваемый каждой транзакции. Если этот код начинается с английской буквы «c», это означает аннулирование. 
2. ***StockCode:*** Код товара (наименования). 5-значный целочисленный номер, уникально присваиваемый каждому отдельному товару. 
3. ***Description:*** Название товара (наименования). 
4. ***Quantity:*** Количество каждого товара (наименования) в одной транзакции. 
5. ***InvoiceDate:*** Дата и время выставления счета. День и время создания транзакции. 
6. ***UnitPrice:*** Цена за единицу. Цена товара за единицу в фунтах стерлингов (£). 
7. ***CustomerID:*** Номер клиента. 5-значный целочисленный номер, уникально присваиваемый каждому клиенту. 
8. ***Country:*** Название страны. Название страны, в которой проживает клиент.
