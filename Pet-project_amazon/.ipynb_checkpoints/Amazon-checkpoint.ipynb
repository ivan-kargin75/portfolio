# Анализ данных

## Создание необходимых дополнительных признаков

```python
# Создадим новый признак (наличие скидки в заказе):
df["HasDiscount"] = (df["Discount"] > 0).astype(int)
```

```python
df.head()
```

## Сравнение показателей для двух групп заказов (со скидкой и без)

```python
# Рассчитаем процент заказов со скидкой:
order_has_discount = len(df[df.HasDiscount == 1]) / len(df) * 100
print(f"Процент заказов со скидкой: {order_has_discount}%")
```

```python
# Сравним средний чек от заказов для двух групп (со скидкой и без):
avg_amount_disc = round(df[df.HasDiscount == 1].TotalAmount.mean(), 1)
avg_amount_no_disc = round(df[df.HasDiscount == 0].TotalAmount.mean(), 1)

# Сравним среднее количества товаров в заказах для двух групп (со скидкой и без):
avg_quantity_disc = round(df[df.HasDiscount == 1].Quantity.mean(), 3)
avg_quantity_no_disc = round(df[df.HasDiscount == 0].Quantity.mean(), 3)

# Сравним доли отмененных заказов для двух групп (со скидкой и без):
cancelled_disc = (df[df.HasDiscount == 1].OrderStatus == "Cancelled").sum()
cancelled_no_disc = (df[df.HasDiscount == 0].OrderStatus == "Cancelled").sum()

# Сравним доли возвращенных заказов для двух групп (со скидкой и без):
returned_disc = (df[df.HasDiscount == 1].OrderStatus == "Returned").sum()
returned_no_disc = (df[df.HasDiscount == 0].OrderStatus == "Returned").sum()
```

```python
compare_groups = pd.DataFrame({"Metrics": ["avg_amount", "avg_quantity", "cancelled", "returned"],
                               "No_discount": [avg_amount_no_disc, avg_quantity_no_disc, cancelled_no_disc, returned_no_disc], 
                               "Discount": [avg_amount_disc, avg_quantity_disc, cancelled_disc, returned_disc]})
compare_groups
```

```python
# Рассчитаем процент отмен и возвратов заказов со скидкой:
discount = df[df["HasDiscount"] == 1].TotalAmount
no_discount = df[df["HasDiscount"] == 0].TotalAmount

per_cancelled = compare_groups.loc[2, "No_discount"] / compare_groups.loc[2, "Discount"] * 100
per_returned = compare_groups.loc[3, "No_discount"] / compare_groups.loc[3, "Discount"] * 100
print(f"От общего числа заказов, процент возвратов и отмен составляет: {round((per_cancelled + per_returned) / 2, 2)}%")
```

__У группы заказов со скидкой средний чек заказа статистически меньше на 100 единиц, среднее количество товаров в одном заказе такое же, а количество отмен и возвратов выше в среднем на 69.37%.__

## Проверка гипотезы о том, что наличие скидки уменьшает сумму покупки.

__Обозначим нулевую и альтернативную гипотезу:__
* H0 - средний чек заказов со скидкой не отличается от среднего чека заказов без скидки. 
* H1 - средний чек заказов со скидкой ниже среднего чека заказов без скидки. 

```python
# Проверим нормальность распределения данных с помощью графика:

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4), sharey=True)
ax1.hist(no_discount)
ax2.hist(discount)
fig.suptitle("Распределения discount и no_discount")
ax1.set_xlabel("Сумма заказа")
ax2.set_xlabel("Сумма заказа")
ax1.set_ylabel("Количество заказов")
ax1.set_title("no_discount")
ax2.set_title("discount");
```

```python
# Проверим нормальность распределения данных с теста Шапиро-Уилка:
print(stats.shapiro(discount.sample(1000)).pvalue)
print(stats.shapiro(no_discount.sample(1000)).pvalue)
```

__Опираясь на построенные графики и pvalue с пороговым значением 0.05, можно сделать вывод: данные не распределены нормально.__

Так как выборки не зависимы и не подчиняются нормальному расределению, используем тест Манна-Уитни.

```python
# Используем тест Манна-Уитни:
stats.mannwhitneyu(discount, no_discount, alternative="less")
```

По результатам теста Манна-Уитни значение pvalue оказалось меньше порогового значения, равного 0.05. Следовательно нулевую гипотезу мы можем отвергнуть.

__Вывод:
Скидка ведет к статистически значимому снижению чека.__

## Считаем выручку для группы заказов со скидкой и без, а также для каждой категории товаров отдельно

```python
# Рассчитаем суммарную выручку от заказов для двух групп (со скидкой и без) без учета стоимости отмененных заказов:
revenue_no_disc = df[((df.OrderStatus != "Cancelled") & (df.OrderStatus != "Returned") & (df["HasDiscount"] == 0))].TotalAmount.sum()
revenue_disc = df[((df.OrderStatus != "Cancelled") & (df.OrderStatus != "Returned") & (df["HasDiscount"] == 1))].TotalAmount.sum()

x = ["revenue_no_disc", "revenue_disc"]
y = [revenue_no_disc, revenue_disc]

print(f"Суммарная выручка для заказов без скидки: {round(y[0])}")
print(f"Суммарная выручка для заказов со скидкой: {round(y[1])}\n")

plt.figure(figsize=(4, 6))
plt.bar(x, y,  color=["orange", "green"])
plt.xlabel("Категории заказов")
plt.ylabel("Выручка")
plt.title("Суммарная выручка для двух групп");
```

```python
revenue_disc / (revenue_disc + revenue_no_disc) * 100
```

```python
revenue_disc / revenue_no_disc * 100
```

__Сумма выручки от заказов со скидкой составила 56.75% общей выручки, что на 31% выше суммы выручик от заказов без скидки.__

```python
# Посчитаем выручку каждой категории для двух групп заказов (со скидкой и без):

print("Суммарная выручка заказов со скидкой и без скидки во всех категориях товаров (без учета стоимости отмененых товаров):\n")
print("no_discount:      discoun:\n")
for category in df.Category.unique():
    
    no_disc = df[(df.OrderStatus != "Cancelled") & (df.OrderStatus != "Returned") & (df["HasDiscount"] == 0) & (df["Category"] == category)].TotalAmount.sum()
    disc = df[(df.OrderStatus != "Cancelled") & (df.OrderStatus != "Returned") & (df["HasDiscount"] == 1) & (df["Category"] == category)].TotalAmount.sum()
    
    no_disc = round(no_disc)
    disc = round(disc)
    
    print("\033[93m", category, "\033[0m")
    print(no_disc, "         ", disc, "\n")
```

__Во всех категориях общая сумма выручки от заказов без скидки уступает заказам, оформленным со скидкой.__

## Расчет показателей для первых покупателей

```python
# Рассчитаем количество клиентов, в первом заказе которых имеется товара со скидкой:
first_order_disc = df.groupby("CustomerID", as_index=False).agg({"OrderDate": "min", "HasDiscount": "first"}).HasDiscount.sum()
first_order_disc

print(f"Количество клиентов, совершивших первый заказ: {df.CustomerID.nunique()}")
print(f"Количество клиентов, совершивших первый заказ товара со скидкой: {first_order_disc}")
```

```python
per_order_disc = round(first_order_disc / df.CustomerID.nunique() * 100, 2)
print(f"От общего числа новых клиентов, {per_order_disc}% из них в первом заказе оформляли товар со скидкой.")
```

```python
x = ["Все покупатели", "Оформлявшие \nпервый заказ\n со скидкой"]
y = [df.CustomerID.nunique(), first_order_disc, ]

plt.figure(figsize=(5, 7))
plt.bar(x, y,color=("orange", "green"))
plt.xlabel("Категории покупателей")
plt.ylabel("Колтчество покупателей")
plt.title("Сравнение двух групп покупателей");
```
